```python
def xyplot(*args, **kwargs):
    data = args[0]
    x = data[0]
    y = data[1]
    
    if "linestyle" not in kwargs:
        kwargs["linestyle"] = "dashed"
        
    if "marker" not in kwargs:
        kwargs["marker"] = "o"
    
    plot(x,y, **kwargs)
```

```python
FEEDS = np.array([750,750,30]) # feeds, per-axis
FEEDS_RAPID = np.array([1500,1500,1500])

class Path(object):
    "A complete three-dimensional path of points"
    def __init__(self, pts):
        self.pts = pts

    def path(self):
        return np.copy(self.pts)
    
    def toplot(self):
        pts = self.path()
        return pts[:,0:2].T # oi
    
    def plot(self, plotfn):
        plotfn(self.toplot())
        
    def gcode_name(self):
        return "Path, %d pts" % len(self.pts)
    
    def extent(self):
        pts = self.path()
        
        minima = [999,999,999]
        maxima = [-999,-999,-99]
        
        for p in pts:
            for i in range(3):
                if p[i] < minima[i]:
                    minima[i] = p[i]
                if p[i] > maxima[i]:
                    maxima[i] = p[i]
                    
                    
        return [ minima, maxima ]
            
    def gcode(self, pt0):
        spindle_speed = 15000
        
        pts = self.path()
        epsilon = 0.01
        
        dt = 0.0
        
        if len(pts) < 1:
            return ""
        
        retval = ""
        
        speed0 = 0
        for pt in pts:
            travel = pt-pt0
            
            feeds_here = FEEDS
            if pt[2] >= 1 and pt0[2] >= 1: # not in work, hopefully...
                feeds_here = FEEDS_RAPID
            
            dp = travel
            wdp = feeds_here * dp
            
            speed = np.sqrt(np.dot(wdp,wdp)/np.dot(dp,dp))
            
            line_dt = np.sqrt(np.dot(dp,dp)) / speed
            
            if not np.isnan(line_dt):
                dt += line_dt
            
            dspeed = speed - speed0
            
            cur_line = "G1 ";
            emitted = False;
            for (axis,delta,value) in zip("XYZF", 
                                          (travel[0],travel[1],travel[2],dspeed), 
                                          (pt[0],pt[1],pt[2],speed)):
                if np.abs(delta) > epsilon:
                    cur_line += "%s%f " % (axis, value)
                    emitted = True
                    
            if emitted:
                retval += cur_line + " (dt %0.3f)\n" % line_dt
                
            pt0 = pt
            speed0 = speed
            
        
        header = ""
        header += "(%s)\n" % self.gcode_name()
        header += "(Extent: %s)" % str(self.extent())
        header += "(Estimated time: %f)\n\n" % dt
        header += "S%d M3\n" % spindle_speed
        footer = "G0 Z15\nM30\n"
        
        return header + retval + footer
    
class PathFlat(Path):
    "A flat path of a given depth"
    def __init__(self, xy_pts, depth):
        self.pts = [ (p[0],p[1],z) for p,z in zip(xy_pts, depth) ]
        self.depth = depth
        
    def gcode_name(self):
        return "PathFlat: depth=%d, %d points" % (self.depth, len(self.pts))
        
class PathArc(Path):
    """Create a partial (flat) circular path, approximated by N segments"""
    def __init__(self, center, radius, depth, theta0_deg, theta1_deg, N=None):
        self.center = center
        self.radius = radius
        self.depth = depth
        self.theta0 = theta0_deg / 180.0 * np.pi
        self.theta1 = theta1_deg / 180.0 * np.pi
        
        self.N = N
        
        n = N
        if None == n:
            n = np.abs(radius * (self.theta1-self.theta0)/(2*np.pi) * 10)
        self.pts = self._pts(n)

    def _pts(self, n):
        thetas = np.linspace(self.theta0, self.theta1, n+1)
        xs = self.center[0] + self.radius*np.cos(thetas)
        ys = self.center[1] + self.radius*np.sin(thetas)
        zs = [ -1*self.depth for _ in ys ]
        
        return zip(xs, ys, zs)

        
class PathCircle(PathArc):
    """Create a complete (flat) circular path, approximated as an N-gon"""
    def __init__(self, center, radius, depth, N=None):
            PathArc.__init__(self, center, radius, depth, 0, 360, N)

            
class PathJoin(Path):
    """Create a sequence of paths"""
    def __init__(self, children=None):
        if None == children:
            self.children = []
        else:
            self.children = children
        
    def add(self, p):
        self.children.append(p)
        
    def path(self):
        retval = []
        for p in self.children:
            retval.extend(p.path())
        return np.array(retval)

class PathJoinClosed(PathJoin):
    def path(self):
        self.real_children = self.children
        pt0 = self.children[0].path()[0]
        pt1 = self.children[-1].path()[-1]
        self.add(Path([pt1,pt0]))
        retval = super(PathJoinClosed,self).path()
        self.children = self.real_children
        return retval
    
class PathJoinSafe(PathJoin):
    def __init__(self, children=None, z_safe=1.0):
        self.z_safe = z_safe
        
        self.children = []
        
        if children:
            for c in children:
                self.add(c)

    def add(self, p):
        pts = p.path()
        if len(pts) < 1:
            return
        
        safe_in = pts[0].copy()
        safe_in[2] = self.z_safe
        
        safe_out = pts[-1].copy()
        safe_out[2] = self.z_safe
        
        self.children.append(Path([safe_in]))
        self.children.append(p)
        self.children.append(Path([safe_out]))
        
class PathTranslate(Path):
    def __init__(self, child, xlate):
        self.xlate = xlate
        self.child = child
        
    def path(self):
        pts = self.child.path()
        return pts + self.xlate
    
class PathRotateAboutZOrigin(Path):
    def __init__(self, child, phi_deg):
        self.child = child
        self.phi = -1 * phi_deg * np.pi / 180.0
        
    def path(self):
        c = np.cos(self.phi)
        s = np.sin(self.phi)
        R = np.array([[c, -s, 0], [s, c, 0], [0,0,1]])
        
        return np.dot(self.child.path(),R)
        
class PathRotateAboutXYPoint(Path):
    def __init__(self, child, center, phi_deg):
        center3 = np.array([center[0], center[1], 0])
        self.pts = PathTranslate(PathRotateAboutZOrigin(PathTranslate(child, center3*-1), phi_deg), center3).path()
        
class PathSpiral(Path):
    def __init__(self, center, radius, z_top, z_bottom, z_step, direction="CCW", N=None):
        self.center = center
        self.radius = radius
        self.z_top = z_top
        self.z_bottom = z_bottom
        self.z_step = z_step
        self.direction = direction
        self.N = N
        
        n = N
        if None == self.N:
            n = int(np.abs(radius * 10))
            
        if n < 4:
            n = 4
        
        if self.direction == "CCW":
            self.theta0 = 0
            self.theta1 = 360
        elif self.direction == "CW":
            self.theta0 = 360
            self.theta1 = 0
        else:
            raise Exception("Unknown direction '%s'" % self.direction)

        
        # numpy.arange is explicitly documented as possibly overflowing, so sanity check.
        zs = [ z for z in np.arange(z_top, z_bottom, z_step*-1.0) if z >= z_bottom ]
        
        # guarantee we get "really close" to the bottom
        if (len(zs) == 0) or (np.abs(zs[-1] - z_bottom) > z_step/10.0):
            zs.append(z_bottom)

        n_steps = len(zs) * n

        thetas = [ 2.0*np.pi*i/n for i in range(n_steps) ]
        if self.direction == "CW":
            thetas = [ t * -1 for t in thetas ]
        
        x_coords = [ self.center[0] + self.radius*np.cos(t) for t in thetas ]
        y_coords = [ self.center[1] + self.radius*np.sin(t) for t in thetas ]
        
        z_coords = [ self.z_top - (self.z_top-self.z_bottom+0.0)*i/n_steps for i in range(n_steps) ]
        
        path_coords = zip(x_coords, y_coords, z_coords)
        
        self.pts = path_coords
        
 #       self.children = [ PathArc(self.center, self.radius, z, self.theta0, self.theta1, self.N) for z in zs ]
        
class PathPeckDrill(Path):
    def __init__(self, center, z_top, z_bottom, z_step):
        self.center = center
        self.z_top = z_top
        self.z_bottom = z_bottom
        self.z_step = z_step
        
        self.z_safe = 1.0
        
        
        center_drill = [ [self.center[0], self.center[1], self.z_safe], [self.center[0], self.center[1], self.z_top] ]
        
        dz = -1.0 * np.abs(self.z_step)
        
        final_z = self.z_safe # the last z position, to make sure we hit bottom
        
        for z in np.arange(self.z_top, self.z_bottom, dz):
            if z >= self.z_bottom:
                center_drill.append( [self.center[0], self.center[1], z ] )
                center_drill.append( [self.center[0], self.center[1], self.z_top ] )
                final_z = z

        if final_z != self.z_bottom:
            center_drill.append( [self.center[0], self.center[1], self.z_bottom] )
            
        center_drill.append( [self.center[0], self.center[1], self.z_safe] )
        
        self.pts = center_drill
        

class PathRoundPocket(PathJoin):
    def __init__(self, radius, z_top, z_bottom, z_step, direction="CCW", N=None, mill_dia=3.2):
        self.center = (0,0)
        self.radius = radius
        self.z_top = z_top
        self.z_bottom = z_bottom
        self.z_step = z_step
        self.direction = direction
        self.N = N
        
        self.mill_dia = 3.2
        self.stepover = 0.4 # fraction of mill_dia

        self.z_safe = 1.0
        
        self.children = []
        
        self.compute_path()
        

    def outward_spiral(self, z):
        retpath = []
        
        r0 = 0
        r_final = self.radius
        dr_per_rev = self.mill_dia * self.stepover
        
        for r in np.arange(r0, r_final, dr_per_rev):
            r_prime = r + dr_per_rev
            if r_prime > r_final:
                r_prime = r_final
                
            n = np.ceil(np.abs(r_prime * 10))
            if n < 4:
                n = 4
                
            while dr_per_rev / n > 0.1:
                n *= 2
                
            r_step = (r_prime - r)/n
            
            for r_i,t_i in zip( np.linspace(r, r_prime, n), np.linspace(0, 2*np.pi, n) ):
                retpath.append([self.center[0] + r_i*np.cos(t_i), self.center[1] + r_i*np.sin(t_i), z])

        n = np.ceil(np.abs(r_final*10))
        if n < 10:
            n = 10
        for t in np.linspace(0, 2*np.pi, n):
            retpath.append([self.center[0]+r_final*np.cos(t), self.center[1] + r_final*np.sin(t), z])
                
        return Path(retpath)
        
    def compute_path(self):
        # drill hole in center
        
        dz = -1.0 * np.abs(self.z_step)
        
        last_z = self.z_top
        for z in np.arange(self.z_top, self.z_bottom, dz):
            if z >= self.z_bottom:
                self.children.append(PathPeckDrill(self.center, last_z, z, self.z_step))
                self.children.append(self.outward_spiral(z))
                

class PathRampDown(Path):
    def __init__(self, child, z_top, z_bottom, z_step):
        self.child = child
        
        z_step = 1.0 * z_step # ensure floating
        
        if z_step > 0:
            z_step = -1.0 * z_step

        base_path = self.child.path()
        n_points = len(base_path)
        
        n_passes = np.ceil((z_bottom-z_top)/z_step)
        
        n = int(n_passes * n_points)
        
        zs = np.linspace(z_top+np.abs(z_step), z_bottom, n)
        xs = [ base_path[i % n_points][0] for i in range(n)]
        ys = [ base_path[i % n_points][1] for i in range(n)]
        
        ramp_spiral = zip(xs,ys,zs)
        
        final_pass = [ [x,y,z_bottom] for x,y,z in base_path ]

        overall_path = []
        overall_path.extend(ramp_spiral)
        overall_path.extend(final_pass)
        
        self.pts = np.array(overall_path)
        
```

```python
thickness = 6.35
```

```python
r_outer = 62.0/2
r_inner = 60.5/2

p_sixth_outer = PathArc((0,0), r_outer, 0, 0, 60, 10)
p_sixth_inner = PathArc((0,0), r_inner, 0, 60, 120, 10)

p_third = PathJoin([p_sixth_outer, p_sixth_inner])

p_third.plot(xyplot)
xlim(-50,50)
ylim(-50,50)
```

```python
p_circle = PathJoinClosed( [p_third, PathRotateAboutZOrigin(p_third, 120), PathRotateAboutZOrigin(p_third, 240)])

p_circle.plot(xyplot)
xlim(-50,50)
ylim(-50,50)
```

```python
p_cutout = PathRampDown(p_circle, 0, -thickness, -1)

p_cutout.plot(xyplot)

xlim(-50,50)
ylim(-50,50)
```

```python
w_ring = 5
w_flat = 5

r_cut = r_outer + w_ring

theta_flat = np.arcsin(w_flat / (2*r_cut)) * 2.0 * 360/(2.0*np.pi)
theta_flat_janky = w_flat / (2*np.pi*r_cut) * 360

p_outer_path = PathJoinClosed([PathRotateAboutZOrigin(PathArc((0,0),r_cut, 0, theta_flat, 120,15), i*120) for i in range(3)])


p_outer_cut = PathRotateAboutZOrigin(PathRampDown(p_outer_path, 0, -thickness, -1), -30)
p_outer_cut.plot(xyplot)

xlim(-50,50)
ylim(-50,50)

theta_flat, theta_flat_janky
```
